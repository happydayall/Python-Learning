```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from ipywidgets import interact
```

```python
x = np.linspace(0 , 2 * np.pi , 500)
```

```python
def draw(n):
    y = np.sin(n *x)
    plt.plot(x , y , lw = 5)  # lw 線的粗度
```

```python
draw(5)
```

```python
interact(draw , n= (1. , 50.))
```
